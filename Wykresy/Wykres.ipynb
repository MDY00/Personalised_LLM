```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

# Definicja danych
data = {
    'Mistral': {'x': 0.7860, 'baseline': 0.3655},
    'Phi-2': {'x': 0.7810, 'baseline': 0.6010},
    'Flan-T5': {'x': 0.7940, 'baseline': 0.5700}
}

# Obliczanie gain
for key, values in data.items():
    data[key]['gain'] = ((values['x'] - values['baseline']) / values['baseline']) * 100
    print(f"Gain dla {key}: {data[key]['gain']}%")

# Tworzenie wykresu
names = list(data.keys())
gains = [values['gain'] for values in data.values()]

plt.bar(names, gains, color='#1f77b4')
plt.ylabel('Przyrost (%)', fontsize=16)
plt.title('CLS vs CLS-P Humicroedit', fontsize=16)
plt.xticks(fontsize=16)
plt.savefig("CLS_HUMI.jpg", dpi=300)

plt.show()
```
