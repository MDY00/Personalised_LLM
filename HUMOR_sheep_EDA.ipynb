```python
df_text['Sum'] = df_text[['x', '1', '2', '3', '4', '5']].sum(axis=1)
```

### more than 2 annotators annotated a tweet

```python
df_text = df_text[df_text['Sum']>=2]
```

```python
df_text.count()
```

```python
# Połączenie DataFrame'ów
merged_df = pd.merge(annotations, df_text, left_on='tweet_id', right_on='id', how='inner')
```

```python
merged_df
```

```python
df = merged_df[['tweet_id','text','session_id', 'vote']]
```

```python
df
```

```python
top_annotators = df['session_id'].value_counts()
```

```python
top_annotators = top_annotators[:100]
```

```python
type(top_annotators)
```

```python
top_annotators
```

```python
top_annotators_df = top_annotators.reset_index()
top_annotators_df.columns = ['session_id', 'count']
```

```python
top_annotators_df['count'].sum()
```

```python
top_annotators_df
```

```python
df['session_id'].nunique()
```

```python
tmp_df = pd.merge(df, top_annotators_df, on='session_id', how='inner')
```

```python
tmp_df = tmp_df.drop(columns=['count'])
```

```python
tmp_df
```

```python
# Przekształcenie DataFrame'a
pivot_df = tmp_df.pivot_table(index=['tweet_id', 'text'], columns='session_id', values='vote', aggfunc='sum', fill_value=0)

# Zresetowanie indeksu
pivot_df = pivot_df.reset_index()

# Wyświetlenie DataFrame'a po przekształceniu
pivot_df[:5]
```

```python
pivot_df = pivot_df.drop(columns=['session_id'])
```
