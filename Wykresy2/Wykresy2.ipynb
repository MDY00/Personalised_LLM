# Humi

```python
import matplotlib.pyplot as plt

# Definicja danych
experiments = ['UserID', 'OneHot', 'HuBiMedium', 'HuBiFormula', 'HuBiSimple']
f1_scores = [0.7786, 0.7567, 0.7774, 0.7582, 0.7535]

# Obliczanie wartoci gain
baseline = 0.5538  # Warto dla 'Baseline'
gains = [(x - baseline) / baseline * 100 for x in f1_scores]

# Tworzenie wykresu
plt.figure(figsize=(8, 6))
bars = plt.bar(experiments, gains, color='#1f77b4')

# Dodawanie wartoci gain na rodku kadego supka
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval/2, round(yval, 2), ha='center', va='center', fontsize=16, color='black')

plt.ylabel('Przyrost [%]', fontsize=16)
plt.title('Wyniki dla osadze modelu BERT na zbiorze Humicroedit', fontsize=16)
plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=16)

# Zapisywanie wykresu jako plik .jpg
plt.savefig("BERT_HUMI.jpg", dpi=300, bbox_inches='tight')

plt.show()
```
