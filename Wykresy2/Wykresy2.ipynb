```python
import matplotlib.pyplot as plt

# Definicja danych
experiments = ['UserID', 'OneHot', 'HuBiMedium', 'HuBiFormula', 'HuBiSimple', 'Baseline']
f1_scores = [0.7786, 0.7567, 0.7774, 0.7582, 0.7535, 0.5538]
errors = [0.0036, 0.0063, 0.0012, 0.0089, 0.0135, 0.0083]

# Tworzenie wykresu
plt.figure(figsize=(10, 6))
bars = plt.bar(experiments, f1_scores, yerr=errors, color='#1f77b4', capsize=10)

# Dodawanie wartości F1 Macro na środku każdego słupka
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval/2, round(yval, 4), ha='center', va='center', fontsize=16, color='black')

plt.ylabel('F1 Macro', fontsize=16)
plt.title('Wyniki dla osadzeń modelu BERT na zbiorze Humicroedit', fontsize=16)
plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=16)
plt.savefig("BERT_HUMI.jpg", dpi=300, bbox_inches='tight')
plt.show()
```
