```python
import os
import pandas as pd

# Inicjalizacja listy do przechowywania artw
jokes = []

# Przechodzenie przez pliki w folderze
folder_path = "Data/Jester/Teksty/jokes"
for i in range(100):
    filename = f"init{i+1}.html"
    file_path = os.path.join(folder_path, filename)
    print(file_path)
    with open(file_path, "r", encoding="utf-8") as file:
        # Odczyt zawartoci pliku
        html_content = file.read()
        # Znalezienie tekstu pomidzy znacznikami <!--begin of joke --> oraz <!--end of joke -->
        start_marker = "<!--begin of joke -->"
        end_marker = "<!--end of joke -->"
        start_index = html_content.find(start_marker)
        end_index = html_content.find(end_marker)
        if start_index != -1 and end_index != -1:
            joke = html_content[start_index + len(start_marker):end_index].strip()
            print(filename)
            print(joke)
            print("--"*20)
            joke = joke.replace('\t', '').replace('</UL>', '').replace('<br>', '').replace('<BR>', '').replace('\n', '').replace('<p>', '').replace('<P>', '').replace('<i>', '').replace('</i>', '')
            jokes.append(joke)

# Zapisanie artw do pliku CSV
output_file_path = "Data/Jester/zartow.csv"
df = pd.DataFrame({"Joke": jokes})
df.to_csv(output_file_path, index=False, encoding="utf-8")

# Opcjonalnie: zapisanie artw do DataFrame'a
df = pd.DataFrame({"Joke": jokes})
print(df)
```

```python
df_jokes = pd.read_csv('Data/Jester/zartow.csv')
```

```python
df_jokes['id_joke'] = df_jokes.index
df_jokes['id_joke'] = df_jokes['id_joke'].apply(lambda x: x + 1)
```

```python

df_annotator_1 = pd.read_csv('Data/Jester/jester-data-1.csv', header=None)
```

```python
user_ids = range(1, len(df_annotator_1) + 1)
df_annotator_1.insert(0, 'annotator_id', user_ids)  # Wstawiamy kolumn 'User_ID' na pierwsz pozycj
```

```python
df_annotator_1 = df_annotator_1.rename(columns={0: 'no_examples'})
```

```python
melted_df = df_annotator_1.melt(id_vars=['annotator_id', 'no_examples'], var_name='joke_id', value_name='grade')
```

```python
merged_df = melted_df.merge(df_jokes, left_on='joke_id', right_on='id_joke')

final_result = merged_df[['joke_id', 'annotator_id', 'no_examples', 'grade', 'Joke']]
```

```python

final_result = final_result.rename(columns={'Joke': 'joke_text'})

final_result = final_result[['joke_id', 'joke_text', 'annotator_id', 'grade','no_examples']]
```

```python

unique_annotator_ids = top_x_df['annotator_id'].nunique()

print("Liczba unikalnych annotator_id:", unique_annotator_ids)
```

```python
top_x_df['is_funny'] = (top_x_df['grade'] > 0).astype(int)
top_x_df['is_not_funny'] = (top_x_df['grade'] <= 0).astype(int)
```

```python
from sklearn.model_selection import train_test_split

train_set, tmp_set = train_test_split(top_x_df, test_size=0.2, stratify=top_x_df['annotator_id'])
```

```python
test_set, val_set = train_test_split(tmp_set, test_size=0.5, stratify=tmp_set['annotator_id'])
```

```python
print(f"Train :{len(train_set)}")
print(f"Val : {len(val_set)}")
print(f"test : {len(test_set)}")
```

```python
# Sprawdzenie
unique_annotators_top_x = set(top_x_df['annotator_id'].unique())
unique_annotators_train = set(train_set['annotator_id'].unique())
unique_annotators_val = set(val_set['annotator_id'].unique())
unique_annotators_test = set(test_set['annotator_id'].unique())

if unique_annotators_top_x == unique_annotators_train == unique_annotators_val == unique_annotators_test:
    print("Yep.")
else:
    print("No")
```
