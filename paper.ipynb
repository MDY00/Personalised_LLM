```python
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd

data = {
    'full': [0.922],
    'similar': [0.865],
    'composite': [0.863],
    'basic': [0.857],
    'unbalanced': [0.842]
}

df = pd.DataFrame(data)

import seaborn as sns
import matplotlib.pyplot as plt

# Ustawienie stylu seaborn na "white" aby usunąć linie siatki
sns.set_style("white")

# Ustawienie palety kolorów na "blue"
sns.set_palette("Blues_r")

# Utworzenie wykresu słupkowego
plt.figure(figsize=(10, 6))
barplot = sns.barplot(data=df)

# Dodanie wartości do słupków
for i, col in enumerate(df.columns):
    barplot.text(i, df[col].values[0]/2, round(df[col].values[0], 3), color='black', ha='center')

plt.title("Mean f1 Score for BERT on test dataset")
plt.ylabel("Mean F1 Score")

# Przekrzywienie etykiet osi x
plt.xticks(rotation=45)

plt.show()
```

```python
data = {
    'full': [0.922],
    'similar': [0.865],
    'composite': [0.863],
    'basic': [0.857],
    'unbalanced': [0.842]
}

# Przekształcenie słownika na DataFrame
df = pd.DataFrame(data)

# Obliczenie wartości baseline
baseline = df['unbalanced'][0]

# Obliczenie nowych wartości i utworzenie nowego DataFrame
new_data = {col: [(x - baseline) / (1 - baseline) for x in df[col]] for col in df.columns}
new_df = pd.DataFrame(new_data)

print(new_df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ustawienie stylu seaborn na "white" aby usunąć linie siatki
sns.set_style("white")

# Ustawienie palety kolorów na "blue"
sns.set_palette("Blues_r")

# Usunięcie kolumny 'unbalanced'
new_df = new_df.drop('unbalanced', axis=1)

# Utworzenie wykresu słupkowego
plt.figure(figsize=(10, 6))
barplot = sns.barplot(data=new_df)

# Dodanie wartości do słupków
for i, col in enumerate(new_df.columns):
    barplot.text(i, new_df[col].values[0]/2, round(new_df[col].values[0], 3), color='black', ha='center')

plt.title("Gain score for BERT on test dataset")
plt.ylabel("Gain score")

# Przekrzywienie etykiet osi x
plt.xticks(rotation=45)

plt.show()
```
